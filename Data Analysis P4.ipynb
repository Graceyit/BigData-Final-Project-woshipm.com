```python
#把list弄成一个str
wordlist_after_jieba = jieba.cut(text_from_file_with_apath, cut_all = True)
wl_space_split = " ".join(wordlist_after_jieba)#添加了这些词以后再转str

wl_space_split= wl_space_split.replace('的', '')
wl_space_split= wl_space_split.replace('你','')
wl_space_split= wl_space_split.replace('如何','')
wl_space_split= wl_space_split.replace('是', '')
wl_space_split= wl_space_split.replace('分析', '')
wl_space_split= wl_space_split.replace('什么', '')
wl_space_split= wl_space_split.replace('从', '')
wl_space_split= wl_space_split.replace('个', '')
wl_space_split= wl_space_split.replace('了', '')
wl_space_split= wl_space_split.replace('和', '')
wl_space_split= wl_space_split.replace('与', '')
wl_space_split= wl_space_split.replace('在', '')
wl_space_split= wl_space_split.replace('做', '')
wl_space_split= wl_space_split.replace('我', '')
wl_space_split= wl_space_split.replace('中', '')
wl_space_split= wl_space_split.replace('有', '')
wl_space_split= wl_space_split.replace('这', '')
wl_space_split= wl_space_split.replace('让', '')
wl_space_split= wl_space_split.replace('到', '')
wl_space_split= wl_space_split.replace('新', '')
wl_space_split= wl_space_split.replace('3', '')
wl_space_split= wl_space_split.replace('为什么', '')
wl_space_split= wl_space_split.replace('之', '')
wl_space_split= wl_space_split.replace('信', '')
wl_space_split= wl_space_split.replace('怎么', '')
wl_space_split= wl_space_split.replace('1', '')
```

```python
#用jieba把词分开
words = wl_space_split.split()
words
```

```python
filepath = 'stopwords.txt'#stopwords把stopwords放进去，然后字数过1的，不在stopwords里面的都算进去
stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
#remove stopwords
processed_word_list = []
for word in words:
    if word not in stopwords and len(word)>1: #remove single word
        processed_word_list.append(word)
```

```python
#词频排序
word_count = pd.Series(processed_word_list).value_counts().sort_values(ascending=False)[0:60]
print(word_count)
```
