### Ranking by the number of favorites

```python
import numpy as np
data=pd.pivot_table(df,values=['favorite'],index='author',aggfunc=[np.sum,np.size])
data['avg']=data[('sum','favorite')]/data[('size','favorite')]
data['avg']=data['avg'].astype('int')
```

```python
data.columns=data.columns.get_level_values(0)
data.columns = ['total_favorites','ariticls_num','avg_favorites']
```

```python
data=data.query('ariticls_num > 4')
```

```python
data_f = data.sort_values(by=['avg_favorites'],ascending=False)
data_f[:10]
```

### 臻龙VS老曹

```python
from pyecharts import Radar
v1 = [[7991,5,1598]]
v2 = [[4012,844,4]]
c_schema= [{"name": "总点赞数", "max": 8000, "min": 1},
           {"name": "文章数", "max": 850, "min": 1},
           {"name": "平均点赞数", "max": 1600, "min": 1}]
radar = Radar()
radar.config(c_schema=c_schema, shape='circle')
radar.add("臻龙", v1, item_color="#f9713c", symbol=None)
radar.add("老曹", v2, item_color="#4e79a7", symbol=None)
radar
```
